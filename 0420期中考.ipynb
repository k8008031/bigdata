```
install.packages("ggplot2")
```

```
library(ggplot2)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1516,color=Position)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1516,color=Position,geom=c("point","smooth"))
```

```
qplot(TotalPoints,data=NBA1516,fill= Position)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1516,geom=c("line"))
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1516,geom = "jitter",alpha = I(1 / 10))
```

```
qplot(FieldGoalsAttempted,TotalPoints,data = NBA1516, facets = Position ~.)
```

```
qplot(ThreesAttempted,data=NBA1516,facet =Position ~. ,binwidth=100)
```
